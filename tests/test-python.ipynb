```python
import pandas as pd
import sys
sys.path.append("..") # Adds higher directory to python modules path.
import os
```

```python
from mapboxgl_py.viz import CircleViz
from mapboxgl_py.utils import df_to_geojson
```

```python
df = pd.read_csv('../examples/points.csv')
df.head(2)
```

```python
data = df_to_geojson(df, ['Avg Medicare Payments', 'Avg Covered Charges', 'Avg Total Payments'])
```

```python
# Open access token from environment variables
viz = CircleViz(os.getenv('MAPBOX_ACCESS_TOKEN'),
            div_id='mapviz', width='100%', height='500px')
```

```python
# Test circle viz
html = viz.createViz(data, "Avg Medicare Payments", 
                     center=[-90, 40], zoom=3, colorStops=[[5000, 'green'],[7500, 'yellow'], [10000, 'red']])
viz.displayViz(html)
```

```python
html = viz.createViz(data, "Avg Covered Charges", 
                     center=[-90, 40], zoom=3, colorStops=[[5000, 'green'],[7500, 'yellow'], [10000, 'red']])
viz.displayViz(html)
```

```python
html = viz.createViz(data, "Avg Total Payments", 
                     center=[-90, 40], zoom=3, colorStops=[[5000, 'green'],[7500, 'yellow'], [10000, 'red']])
viz.displayViz(html)
```
