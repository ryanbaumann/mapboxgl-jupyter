```python
import pandas as pd
import sys
sys.path.append("..") # Adds higher directory to python modules path.
import os
```

```python
from mapboxgl_py.viz import CircleViz
from mapboxgl_py.utils import *
from jenks import jenks
```

```python
df = pd.read_csv('../examples/points.csv')
df['Avg Total Payments'] = df['Avg Total Payments'].round(2)
df['Avg Covered Charges'] = df['Avg Covered Charges'].round(2)
df['Total Discharges'] = df['Total Discharges'].round(2)
df['Avg Medicare Payments'] = df['Avg Medicare Payments'].round(2)
df.head(2)
```

```python
data = df_to_geojson(df, 
                     properties=['Avg Medicare Payments', 'Avg Covered Charges'],
                     precision=4)
```

```python
# Open access token from environment variables
viz = CircleViz(os.getenv('MAPBOX_ACCESS_TOKEN'),div_id='mapviz')
```

```python
# Generate data breaks and color stops from colorBrewer
breaks = jenks(df['Avg Medicare Payments'].tolist(), 8)
colorStops = createColorStops(breaks, colors='YlGnBu')
```

```python
# Test the circle viz on the Avg Medicare Payments property
viz.createViz(data, 
              colorProperty="Avg Medicare Payments", 
              center=[-90, 40], 
              zoom=3, 
              colorStops=colorStops)
```

```python
# Generate data breaks and color stops from colorBrewer
breaks = jenks(df['Avg Covered Charges'].tolist(), 8)
colorStops = createColorStops(breaks, colors='YlOrRd')
```

```python
# Test new breaks and data property
viz.createViz(data, 
              colorProperty="Avg Covered Charges", 
              center=[-90, 40], 
              zoom=3, 
              colorStops=colorStops)
```
